```python
import os
import webbrowser
import datetime
import wikipedia
import pyjokes
import speech_recognition as sr
from gtts import gTTS
import playsound
from IPython.display import display, Javascript
from google.colab import output

# Código Javascript para solicitar permissão do microfone e gravar áudio
RECORD = """
const sleep  = time => new Promise(r => setTimeout(r, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsText(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  const blob = new Blob(chunks)
  text = await b2text(blob)
  resolve(text)
})
"""

display(Javascript(RECORD))

# Função para capturar áudio (adaptada para o Colab)
def get_audio_colab():
    tentativas = 3  # Número máximo de tentativas de captura
    while tentativas > 0:
        try:
            with output.capture():
                said = eval(Javascript('record(5000)'))
            if said:  # Verifica se algo foi capturado
                print(f"Você disse: {said}")
                return said.lower()
            else:
                print("Nenhum áudio detectado. Tentando novamente...")
                tentativas -= 1  # Decrementa as tentativas
        except Exception as e:
            print(f"Erro na captura de áudio: {e}")
            tentativas -= 1  # Decrementa as tentativas
            print(f"Tentativas restantes: {tentativas}")
            if tentativas == 0:
                print("Falha na captura de áudio após várias tentativas.")
                return "" # Retorna string vazia se houver falha após várias tentativas

    return ""  # Retorna string vazia se o microfone não funcionar após várias tentativas

def speak(text):
    tts = gTTS(text=text, lang="en")
    filename = "voice.mp3"
    try:
        os.remove(filename)
    except OSError:
        pass
    tts.save(filename)
    playsound.playsound(filename)

#function to respond to commands
def respond(text):
    print("Text from get audio " + text)
    if "bom dia" in text:
        speak("olá, bom dia!")

    elif 'youtube' in text:
      speak("Você quer buscar pelo que no Youtube?")
      keyword = get_audio()
      if keyword!= '':
          url = f"https://www.youtube.com/results?search_query={keyword}"
          webbrowser.get().open(url)
          speak(f"Aqui está o que encontrei para {keyword}.")
    elif 'hora' in text:
        StrTime = datetime.datetime.now().strftime("%H:%M")
        print(StrTime)
        speak(StrTime)
    elif 'play music' in text or 'play song' in text:
        speak("Now playing...")
        music_dir = "C:\\Users\\My\\Music\\"
        songs = os.listdir(music_dir)
        #counter = 0
        print(songs)
        playmusic(music_dir + "\\" + songs[0])
    elif 'stop music' in text:
        speak("Stopping playback.")
        stopmusic()
    elif 'exit' in text:
        speak("Goodbye, till next time")
        exit()
    elif 'pesquisar' in text:
        speak("O que você gostaria de pesquisar?")
        search = get_audio_colab()
        url = f"https://www.google.com/search?q={search}"
        webbrowser.get().open(url)
    elif 'search' in text:
        speak("Buscando no Wikipedia...")
        query = text.replace("search", "")
        try:
            result = wikipedia.summary(query, sentences=3)
            speak("De acordo com a Wikipedia")
            print(result)
            speak(result)
        except wikipedia.exceptions.PageError:
            speak("Não encontrei essa página na Wikipedia.")
        except wikipedia.exceptions.DisambiguationError as e:
            speak("A Wikipedia encontrou várias páginas para sua busca. Seja mais específico.")
            print(e)
    elif "joke" in text:
        speak(pyjokes.get_joke())
    elif "ate mais" in text:
        speak("Até a próxima!")
        exit()
    elif text == "":
        pass
#play music
def playmusic(song):
    mixer.init()
    mixer.music.load(song)
    mixer.music.play()

#stop music
def stopmusic():
    mixer.music.stop()

while True:
    print("Estou ouvindo...")
    text = get_audio_colab()
    respond (text)
```

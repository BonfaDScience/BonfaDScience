Copiando as imagens de bikes e elefentes do drive para cá

```python
import os
import shutil

# Diretórios de origem (Drive)
source_dirs = {
    "bicycles": "/content/drive/MyDrive/YOLO/images/bicycles",
    "elephants": "/content/drive/MyDrive/YOLO/images/elephants"
}

# Diretórios de destino (local)
dest_dirs = {
    "bicycles": "/content/images/bicycles",
    "elephants": "/content/images/elephants"
}

# Criando os diretórios de destino se não existirem
for key in dest_dirs:
    os.makedirs(dest_dirs[key], exist_ok=True)

# Copiando as imagens de cada categoria
for category, source_dir in source_dirs.items():
    dest_dir = dest_dirs[category]

    # Verificando se o diretório de origem existe antes de tentar copiar
    if not os.path.exists(source_dir):
        print(f"⚠️ Diretório de origem não encontrado: {source_dir}")
        continue

    # Copiando os arquivos
    for filename in os.listdir(source_dir):
        if filename.endswith(".jpg"):  # Apenas arquivos .jpg
            src_file = os.path.join(source_dir, filename)
            dest_file = os.path.join(dest_dir, filename)
            shutil.copy(src_file, dest_file)

    print(f"📂 Imagens copiadas de {source_dir} para {dest_dir}")

```

extraindo as imagens de elefantes e bikes, separando em grupos de treino, teste e validação cada um deles e colocando-os juntos nos mesmos diretorios.

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Caminhos dos diretórios de imagens e labels
elephant_dir = "/content/images/elephants"  # Diretório com as imagens de elefantes
bicycle_dir = "/content/images/bicycles"  # Diretório com as imagens de bicicletas
images_dir = "/content/images"  # Diretório de imagens para treino/teste/validação
labels_dir = "/content/labels"  # Diretório de labels YOLO

# Diretórios de destino para treino, validação e teste
train_img_dir = os.path.join(images_dir, "train")
val_img_dir = os.path.join(images_dir, "val")
test_img_dir = os.path.join(images_dir, "test")
train_label_dir = os.path.join(labels_dir, "train")
val_label_dir = os.path.join(labels_dir, "val")
test_label_dir = os.path.join(labels_dir, "test")

# Garantindo que os diretórios de destino existam
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(val_img_dir, exist_ok=True)
os.makedirs(test_img_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)
os.makedirs(test_label_dir, exist_ok=True)

# Obtendo todas as imagens de elefantes e bicicletas
elephant_images = [os.path.join(elephant_dir, f) for f in os.listdir(elephant_dir) if f.endswith(".jpg")]
bicycle_images = [os.path.join(bicycle_dir, f) for f in os.listdir(bicycle_dir) if f.endswith(".jpg")]

# Separando as imagens de elefantes em treino, validação e teste
train_elephant_files, val_test_elephant_files = train_test_split(elephant_images, train_size=0.7, random_state=42)
val_elephant_files, test_elephant_files = train_test_split(val_test_elephant_files, test_size=0.5, random_state=42)

# Separando as imagens de bicicletas em treino, validação e teste
train_bicycle_files, val_test_bicycle_files = train_test_split(bicycle_images, train_size=0.7, random_state=42)
val_bicycle_files, test_bicycle_files = train_test_split(val_test_bicycle_files, test_size=0.5, random_state=42)

# Função para copiar as imagens e os labels
def copy_files(files, target_img_dir, target_label_dir, source_img_dir, source_label_dir):
    for file in files:
        # Copiar a imagem
        shutil.copy(file, target_img_dir)

        # Caminho do label correspondente
        label_file = file.replace(source_img_dir, source_label_dir).replace(".jpg", ".txt")

        # Verificando se o label existe e copiando para o diretório de labels
        if os.path.exists(label_file):
            shutil.copy(label_file, target_label_dir)

# Copiando as imagens e labels de elefantes para as pastas de treino, validação e teste
copy_files(train_elephant_files, train_img_dir, train_label_dir, elephant_dir, labels_dir)
copy_files(val_elephant_files, val_img_dir, val_label_dir, elephant_dir, labels_dir)
copy_files(test_elephant_files, test_img_dir, test_label_dir, elephant_dir, labels_dir)

# Copiando as imagens e labels de bicicletas para as pastas de treino, validação e teste
copy_files(train_bicycle_files, train_img_dir, train_label_dir, bicycle_dir, labels_dir)
copy_files(val_bicycle_files, val_img_dir, val_label_dir, bicycle_dir, labels_dir)
copy_files(test_bicycle_files, test_img_dir, test_label_dir, bicycle_dir, labels_dir)

print("📂 Imagens e labels de elefantes e bicicletas separados com sucesso!")
```

a conversão COCO para YOLO está se apresentando um problema. O !yolo task=convert, não está funcionando corretamente. Vou tentar convertar as anotações JSON para .txt na mão.

```python
# Converte as anotações JSON em anotações .txt
import json
import os

# Caminho do arquivo COCO JSON
coco_json = "/content/annotations/instances_train2017.json"
# Pasta onde as imagens de treino estão localizadas
image_dir = "/content/images/train"
# Pasta onde os arquivos de label YOLO serão salvos
label_dir = "/content/labels/train"

# Criando diretório de labels se não existir
os.makedirs(label_dir, exist_ok=True)

# Carregando anotações do COCO
with open(coco_json, "r") as f:
    coco_data = json.load(f)

# Criando um dicionário que mapeia IDs de imagens para seus nomes
image_id_to_filename = {img["id"]: img["file_name"] for img in coco_data["images"]}

# Processando anotações
for annotation in coco_data["annotations"]:
    image_id = annotation["image_id"]
    category_id = annotation["category_id"] - 1  # Ajuste se necessário
    bbox = annotation["bbox"]

    # COCO usa [x_min, y_min, width, height]
    x_min, y_min, width, height = bbox

    # Normalizando coordenadas para formato YOLO
    img_filename = image_id_to_filename.get(image_id, None)
    if img_filename is None:
        continue

    img_path = os.path.join(image_dir, img_filename)
    if not os.path.exists(img_path):
        print(f"⚠️ Imagem {img_filename} não encontrada!")
        continue

    img_width, img_height = 640, 640  # pode ser ajustado

    x_center = (x_min + width / 2) / img_width
    y_center = (y_min + height / 2) / img_height
    width = width / img_width
    height = height / img_height

    # Criando nome do arquivo de label
    label_filename = os.path.splitext(img_filename)[0] + ".txt"
    label_path = os.path.join(label_dir, label_filename)

    # Salvando a anotação no formato YOLO
    with open(label_path, "a") as label_file:
        label_file.write(f"{category_id} {x_center} {y_center} {width} {height}\n")

print("✅ Conversão concluída! Agora os rótulos estão no formato YOLO.")
```
